# Solution code

```python
def solutionA(A):               # argument A is an array of numbers

    for i in range(len(A)):
        A[i] = abs(A[i])        # convert all elements to their absolute values
        
    A.sort()                   # sort the array (make duplicates from abs conversion adjacent to each other e.g. -3,-2,3 -> 2,3,3)
                               # also allows this code to solve arrays that aren't pre-arranged
    
    count = 0
    for i in range(len(A)):            # for every element in the newly sorted array
        if i == 0 or A[i] != A[i-1]:   # if the current element is the first element (1st is always new) or different from the previous one (new unique number)
            count += 1                 # add to distinct count
            
    return count                 # return the number of distinct numbers in the sorted array
```

# Implementation/Testing proof (Local)

```python
arr1 = [-5, -3, -1, 0, 3, 6]      # exact array from the codility question (should output 5)
print(solutionA(arr1))            # call solution function

# ↓↓ The file should come displaying 5 under this cell as proof that it ran successfully, but feel free to test further by changing the array and rerunning this cell 
```
